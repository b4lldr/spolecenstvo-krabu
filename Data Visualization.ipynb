```python
from matplotlib import pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
names = []
mes_total = []
mes_2016 = []
mes_2017 = []
mes_2018 = []
mes_2019 = []
mes_2020 = []

# names and total
file = open("data/zpravy_podle_jmena_celkem", "r")
lines = file.readlines()
for line in lines:
    names.append(line.split('\t')[1])
    mes_total.append(int(line.split('\t')[0]))
file.close()

# 2016
file = open("data/zpravy_podle_jmena_2016", "r")
lines = file.readlines()
for line in lines:
    mes_2016.append(int(line.split('\t')[0]))
file.close()
# 2017
file = open("data/zpravy_podle_jmena_2017", "r")
lines = file.readlines()
for line in lines:
    mes_2017.append(int(line.split('\t')[0]))
file.close()
# 2018
file = open("data/zpravy_podle_jmena_2018", "r")
lines = file.readlines()
for line in lines:
    mes_2018.append(int(line.split('\t')[0]))
file.close()
# 2019
file = open("data/zpravy_podle_jmena_2019", "r")
lines = file.readlines()
for line in lines:
    mes_2019.append(int(line.split('\t')[0]))
file.close()
# 2020
file = open("data/zpravy_podle_jmena_2020", "r")
lines = file.readlines()
for line in lines:
    mes_2020.append(int(line.split('\t')[0]))
file.close()

# plotting
fig = plt.figure(1, figsize=(48, 8), dpi=200)
chart1 = fig.add_subplot(121)

## bar creating
plt.bar(names, mes_2016, label="2016")

plt.bar(names, mes_2017, bottom=mes_2016, label="2017")

tmp = list(map(lambda x,y: x+y, mes_2016,mes_2017))
plt.bar(names, mes_2018, bottom=tmp, label="2018")

tmp = list(map(lambda x,y: x+y, tmp,mes_2018))
plt.bar(names, mes_2019, bottom=tmp, label="2019")

tmp = list(map(lambda x,y: x+y, tmp,mes_2019))
plt.bar(names, mes_2020, bottom=tmp, label="2020")

## plot info
plt.title("Zprvy napsan danm lovkem", fontsize=20)
plt.xticks(rotation=80)
plt.legend(loc="upper right", fontsize=15)
plt.tick_params(labelsize=20)

### number of messages accurately in the plot
for i, v in enumerate(mes_total):
    plt.text(i - 0.35, v + 100, str(v), fontsize=10)

#plt.savefig('jmena.png', bbox_inches='tight')

plt.show()
```

```python
names = []
messages = []

file = open("data/zpravy_podle_jmena_celkem", "r")
lines = file.readlines()
flag = 0
for line in lines:
    value = int(line.split('\t')[0])
    if value > 1000:
        names.append(line.split('\t')[1])
        messages.append(value)
    elif flag == 0:
        names.append("< 1000 zprv [1]")
        messages.append(value)
        flag = 1
    else:
        flag += 1
        names[-1] = "< 1000 zprv " + "[" + str(flag) + "]"
        messages[-1] += + value
        
file.close()

fig = plt.figure(1, figsize=(48, 8), dpi=200)
chart1 = fig.add_subplot(121)

explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)

plt.pie(messages, labels=names, autopct='%1.1f%%', startangle=90, pctdistance=0.85, explode = explode)

plt.title("Zprvy napsan danm lovkem 2", fontsize=20)

# draw circle in the middle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle) # Equal aspect ratio ensures that pie is drawn as a circle 
plt.tight_layout()

#plt.savefig('jmena_pie.png', bbox_inches='tight')

plt.show()
```

```python
months = []
mes_total = []

file = open("data/zpravy_za_mesic", "r")
lines = file.readlines()
for line in lines:
    months.append(line.split('\t')[1])
    mes_total.append(int(line.split('\t')[0]))
file.close()

fig = plt.figure(1, figsize=(48, 8), dpi=200)
chart1 = fig.add_subplot(121)

plt.bar(months, mes_total)

plt.title("Zprvy napsan v jednotlivch mscch", fontsize=20)
plt.xticks(rotation=80)
plt.tick_params(labelsize=15)

#plt.savefig('mesice.png', bbox_inches='tight')

plt.show()
```
